### BoxPlot

This notebook is based on DataCamp Data Visualization course on Boxplot using Matplotlib.
https://campus.datacamp.com/courses/introduction-to-data-visualization-with-matplotlib/quantitative-comparisons-and-statistical-visualizations?ex=10

```python
import pandas as pd
import re
import matplotlib.pyplot as plt
df = pd.read_csv(r'./data/summer2016.csv')
df.head()
mens_rowing = df[df['Sport'] == 'Rowing']
mens_gymnastics = df[df['Sport'] == 'Gymnastics']
```

```python
fig, ax = plt.subplots()

# Add a boxplot for the "Height" column in the DataFrames
ax.boxplot([mens_rowing["Height"],mens_gymnastics["Height"]])

# Add x-axis tick labels:
ax.set_xticklabels(["Rowing", "Gymnastics"])

# Add a y-axis label
ax.set_ylabel("Height (cm)")

plt.style.use('ggplot')
#ax.legend()
#fig.tight_layout()
plt.savefig("Gym_rowing_weight_boxplot.png")
_ = plt.show()
```
