### Histograms

This notebook is based on DataCamp Data Visualization course on Histogram using Matplotlib.

```python
import pandas as pd
import re
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'./data/summer2016.csv')
df.head()
mens_rowing = df[df['Sport'] == 'Rowing']
mens_gymnastics = df[df['Sport'] == 'Gymnastics']
```

```python

fig, ax = plt.subplots(figsize = (20,8))
# Plot a histogram of "Weight" for mens_rowing
# histtype will change apperance of the bars so that underlying is also visible and change bins from default 10 to 5
ax.hist(mens_rowing['Weight'], label = "Rowing",  histtype = 'step', bins = 5)

# Compare to histogram of "Weight" for mens_gymnastics
ax.hist(mens_gymnastics['Weight'], label = "Gymnastics", histtype = 'step', bins = 5)

# Set the x-axis label to "Weight (kg)"
ax.set_xlabel("Weight (kg)")

# Set the y-axis label to "# of observations"
ax.set_ylabel("# of observations")
ax.set_title(" Weight comaprision of Mens rowing and Gymnastics players")

plt.style.use('ggplot')
ax.legend()
#fig.tight_layout()
plt.savefig("Gym_rowing_weight_historgram.png")
_ = plt.show()
```
