### BarPlot

This notebook is based on DataCamp Data Visualization course on Barplot using Matplotlib.
https://campus.datacamp.com/courses/introduction-to-data-visualization-with-matplotlib/quantitative-comparisons-and-statistical-visualizations?ex=1

```python
import pandas as pd
import re
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'./data/medals_by_country_2016.csv', index_col=0)
df.head()
```

```python
fig, ax = plt.subplots(figsize = (20,8))
ax.bar(df.index, df['Gold'])
#ax.set_xticklabels(df.index, rotation = 90)
# to avoid warning UserWarning: FixedFormatter should only be used together with FixedLocator
ax.tick_params(axis = 'x', rotation = 90) 
ax.set_xlabel("Country")
ax.set_ylabel("Number of Medals")
ax.set_title(" Number of Gold Medals Received by Each Country")
plt.style.use('ggplot')
#fig.tight_layout()
plt.savefig("Medals_barplot.png")
_ = plt.show()
```

### Stacked BarPlot
Now we will add information about all medals i.e. silver and bronze by creating stacked bar chart
We will add method bottom

```python
fig, ax = plt.subplots(figsize = (20,8))
ax.bar(df.index, df['Gold'], label = 'Gold')
ax.bar(df.index, df['Silver'], bottom=df['Gold'], label = 'Silver')
# For mutiple stacks we have to tell which bars will be behind it and this means bottom will be sum of 
# Gold and Silver
ax.bar(df.index, df['Bronze'], bottom=df['Silver'] + df['Gold'], label = 'Bronze') 
#ax.set_xticklabels(df.index, rotation = 90)
# to avoid warning UserWarning: FixedFormatter should only be used together with FixedLocator
ax.tick_params(axis = 'x', rotation = 90) 
ax.set_xlabel("Country")
ax.set_ylabel("Number of Medals")
ax.set_title(" Number of Gold Medals Received by Each Country")
plt.style.use('ggplot')
ax.legend()
#fig.tight_layout()
plt.savefig("Medals_stacked_barplot.png")
_ = plt.show()
```
